```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread("image.png", cv2.IMREAD_GRAYSCALE)

kernels = {
    "M1 (Sobel Horizontal)": np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]),
    "M2 (Sobel Vertical)": np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]]),
    "M3 (Laplacian)": np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]),
    "M4 (Laplacian v2)": np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]]),
    "M5 (Prewitt-like)": np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]]),
    "M5 (Gaussian Blur)": (1/256) * np.array([[1, 4, 6, 4, 1], 
                                              [4, 16, 24, 16, 4], 
                                              [6, 24, 36, 24, 6], 
                                              [4, 16, 24, 16, 4], 
                                              [1, 4, 6, 4, 1]])
}


fig, axes = plt.subplots(2, 4, figsize=(12, 8))
axes = axes.ravel()
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis("off")

for i, (name, kernel) in enumerate(kernels.items()):
    processed_image = cv2.filter2D(image, -1, kernel)
    axes[i+1].imshow(processed_image, cmap='gray')
    axes[i+1].set_title(name)
    axes[i+1].axis("off")
    

plt.tight_layout()
plt.show()
```
